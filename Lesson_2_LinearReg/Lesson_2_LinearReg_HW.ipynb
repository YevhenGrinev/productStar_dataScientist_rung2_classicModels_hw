```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./train.csv')
```

```python
df.head()
```

```python
df.fillna(0, inplace=True)
```

```python
df.head()
```

```python
columns = ['price_doc', 'timestamp', 'full_sq', 'life_sq', 'floor', 'max_floor', 'material', 'build_year', 'num_room', 'kitch_sq', 'state', 'product_type', 'sub_area']
df = df[columns]
df.head()
```

```python
df.dtypes
```

```python
df = df._get_numeric_data()
df.head()
```

price_doc : цена продажи (это целевая переменная)

```python
# With random_state=0 , we get the same train and test sets across different executions.
df_test = df.sample(1000, random_state=0)
df_train = df.drop(index=df_test.index)
# df_test.index

df_test = df_test.reset_index(drop=True)
df_train = df_train.reset_index(drop=True)
```

```python
X_train = df_train[df_train.columns[1:]]
# X_train.head()
y_train = df_train[df_train.columns[:1]]
# y_train.head()

X_test = df_test[df_test.columns[1:]]
# X_test.head()
y_test = df_test[df_test.columns[:1]]
# y_test.head()
```

```python
from sklearn.metrics import mean_squared_log_error
```

```python
predictions = np.array([y_train.median()]*1000)
# predictions
```

```python
print(f'RMSLE: {np.sqrt(mean_squared_log_error(y_test, predictions))}')
```

```python
baseline = np.sqrt(mean_squared_log_error(y_test, predictions))
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
min_max = MinMaxScaler()
standard = StandardScaler()

# In a nutshell: fitting is equal to training. Then, after it is trained, the model can be used to make predictions, 
# usually with a .predict() method call.
min_max.fit(X_train)
standard.fit(X_train)

# Transformers are classes that enable data transformations while preprocessing the data for machine learning. 

X_train_min_max = min_max.transform(X_train)
X_test_min_max = min_max.transform(X_test)

X_train_standard = standard.transform(X_train)
X_test_standard = standard.transform(X_test)
```

```python
import sklearn.linear_model as lm
```

```python
# Scaling transforms a set of variables into a new set of variables that have the same order of magnitude
linear_regression = lm.LinearRegression(normalize=False).fit(X=X_train_min_max, y=y_train)
```

```python
def calculate_metric(X_train, X_test, y_train, y_test, model):
    
    predictions_train = model.predict(X_train)
    predictions_test = model.predict(X_test)
    
# numpy.where() function returns the indices of elements in an input array where the given condition is satisfied.
    
    if (predictions_train < 0).any() or (predictions_test < 0).any(): 
        predictions_train[np.where(predictions_train < 0)] = 0
        predictions_test[np.where(predictions_test < 0)] = 0
    
    metric_test = np.sqrt(mean_squared_log_error(y_test, predictions_test))
    metric_train = np.sqrt(mean_squared_log_error(y_train, predictions_train))
    
    delta_baseline = baseline - metric_test
    
    print(f'RMSLE test {metric_test}')
    print(f'RMSLE train {metric_train}')
    print(f'Improving metric {delta_baseline}')    
```

```python
calculate_metric(X_train=X_train_min_max, X_test=X_test_min_max, y_train=y_train, y_test=y_test, model=linear_regression)
```

```python
linear_regression_standard = lm.LinearRegression().fit(X=X_train_standard, y=y_train)
```

```python
calculate_metric(X_train=X_train_standard, X_test=X_test_standard, y_train=y_train, y_test=y_test, model=linear_regression_standard)
```

```python
linear_regression_new = lm.LinearRegression().fit(X=X_train_min_max_new, y=y_train_new)
```

```python
calculate_metric(X_train=X_train_min_max_new, X_test=X_test_min_max_new, y_train=y_train_new, y_test=y_test_new, model=linear_regression_new)
```
